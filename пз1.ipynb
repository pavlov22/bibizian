```python
import json
import requests
import matplotlib.pyplot as plt

# Запит до НБУ (приклад — з 17 по 21 березня 2025 року)
url = "https://bank.gov.ua/NBU_Exchange/exchange_site?start=20250317&end=20250321&valcode=eur&json"
nbu_response = requests.get(url)

# Конвертація відповіді в список словників
converted_response = json.loads(nbu_response.content)

# Підготовка даних для графіка
dates = []
rates = []

print("Курси євро за обраний тиждень:")
for item in converted_response:
    date = item["exchangedate"]
    rate = item["rate"]
    print(f"{date}: {rate}")
    dates.append(date)
    rates.append(rate)

# Побудова графіка
plt.figure(figsize=(10, 5))
plt.plot(dates, rates, marker='o', linestyle='-', color='blue')
plt.title("Курс євро (EUR) за тиждень")
plt.xlabel("Дата")
plt.ylabel("Курс (UAH)")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
